### 3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
#getting the google images link
driver = webdriver.Chrome(r"C:/Users/atoranga/Documents/chromedriver.exe")
url='https://images.google.com/'

#asserting search items in search bar and clicking search
def search_key(key):
    driver.get(url)
    search = driver.find_element_by_xpath('//input[@class="gLFyf gsfi"]')
    search.send_keys(key)

    search_btn = driver.find_element_by_xpath('//div[@class="zgAlFc"]')
    search_btn.click()
    time.sleep(3)

#creating empty list to scrape the data
fruits=[]
cars=[]
Machine_Learning=[]
Guitar=[]
Cakes=[]

#calling individual function for each item
search_key('fruits')
time.sleep(2)
#getting image tags and extracting only image link to the lists
img_tag= driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
for i in img_tag[:10]:
    fruits.append(i.get_attribute('src'))
    
search_key('cars')
time.sleep(2)
img_tag= driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
for i in img_tag[:10]:
    cars.append(i.get_attribute('src'))
    
search_key('Machine Learning')
time.sleep(2)
img_tag= driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
for i in img_tag[:10]:
    Machine_Learning.append(i.get_attribute('src'))
    
search_key('Guitar')
time.sleep(2)
img_tag= driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
for i in img_tag[:10]:
    Guitar.append(i.get_attribute('src'))
    
search_key('Cakes')
time.sleep(2)
img_tag= driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
for i in img_tag[:10]:
    Cakes.append(i.get_attribute('src'))

driver.close()

len(fruits),len(cars),len(Machine_Learning), len(Guitar), len(Cakes)

#creating dataframe from the list of data scraped
df = pd.DataFrame({'Fruits':fruits,'Cars':cars,'Machine Learning':Machine_Learning,'Guitar':Guitar,'Cakes':Cakes})
df

```
