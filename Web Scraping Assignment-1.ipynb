
### 1) Write a python program to display all the header tags from wikipedia.org.

```python
#import all necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

#defining a python function which takes url as parameter
def Wiki(url):
    #send the request access to server to get content
    page = requests.get(url)
    #download all the content
    soup = BeautifulSoup(page.content)
    
    #scraping the headers tags
    print("The header tags are:\n")
    for tag in soup.find_all(['h1','h2','h3','h4','h5','h6']):
        print(tag.name,':-',tag.text.strip())
```

```python
#get the link from user
link='https://en.wikipedia.org/wiki/Main_Page'
#call the function
Wiki(link)
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
###      and make data frame

```python
#importing all libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def IMDB_rating(url):
    #requesting the servers through get request.
    page = requests.get(url)
    #download all the content of webpage
    soup = BeautifulSoup(page.content)
    #scraping the top 100 data
    #Name of movie
    Name=[]
    lst=  soup.find_all('td',class_="titleColumn")
    for x in lst:
        for y in x.find_all('a'):
            Name.append(y.text)
            
        
    #IMDB rating of movie
    Rating=[]
    for i in soup.find_all('td', class_="ratingColumn imdbRating"):
        Rating.append(i.text.replace('\n',' '))
    
    #year of release
    YOR=[]
    for i in soup.find_all('span',class_="secondaryInfo"):
        YOR.append(i.text)
           
    
    #creating Dataframe
    y=list(range(1,251))
    df = pd.DataFrame({'Name':Name,'Rating':Rating,'YOR':YOR},index=y)
    df.to_excel('IMDB.xlsx')
        
    #printing top 100 movies
    return df.head(100)
        
```

```python
url = ('https://www.imdb.com/chart/top')
IMDB_rating(url)
```

### 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
### release) and make data frame.

```python
#import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
def IMDB_indian(url):
    #get the request from page
    page = requests.get(url)
    #download all the content of page
    soup= BeautifulSoup(page.content)
    
    #scrape the data of top 100 indian movies
    Name=[]
    lst= soup.find_all('td',class_="titleColumn")
    for i in lst:
        for j in i.find_all('a'):
            Name.append(j.text)
            
    Rating=[]
    for i in soup.find_all('td',class_="ratingColumn imdbRating"):
        Rating.append(i.text.replace('\n',''))
    
    YOR=[]
    for i in soup.find_all('span',class_="secondaryInfo"):
        YOR.append(i.text)
        
    y=list(range(1,251))
    df = pd.DataFrame({'Name':Name,'Rating':Rating,'YOR':YOR}, index=y)

    return df.head(100)
```

```python
url='https://www.imdb.com/india/top-rated-indian-movies/'
IMDB_indian(url)
```

### 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
#import all libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
 
def ICC_menteam(url):
    #open the link using urllib method
    page = requests.get(url)
    
    #get the content to Beautifulsoup
    soup=BeautifulSoup(page.content)
    
    print(soup.title.text)
    table=soup.find('table',class_="table")

    #scrape the data of team name
    Team=[] 
    for i in table.find_all('span',class_="u-hide-phablet"):
        Team.append(i.text)
    
    #scrape positions of all teams
    POS=[] 
    i =table.find('td',class_="rankings-block__banner--pos")
    POS.append(i.text)
    for i in table.find_all('td',class_="table-body__cell table-body__cell--position u-text-right"):
        POS.append(i.text)
    
    #scrape number of matches played
    Matches=[]
    M=[]
    i = table.find('td',class_="rankings-block__banner--matches")
    Matches.append(i.text)

    #scrape points of each team
    Points=[] 
    P=[]
    i = table.find('td',class_="rankings-block__banner--points")
    Points.append(i.text)

    for i in table.find_all('td',class_="table-body__cell u-center-text"):
        
        P.append(i.text)
        M.append(i.text)

    Points= Points + P[1::2]
    Matches=Matches+M[0::2]

    #converting all data into dataframe
    df = pd.DataFrame({'POS':POS,'Team':Team, 'Matches':Matches,'Points':Points},index=POS)
    return df.head(10)
```

```python
link='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
ICC_menteam(link)
```

```python
#import necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def ICC_menbatting(url):
    #downloading content of ICC mens ranking page
    page= requests.get(url)
    page

    soup=BeautifulSoup(page.content)
    #print the title
    print(soup.title.text)

    table = soup.find('table',class_="table rankings-table")

    #scrape the name of player
    Name=[]
    i=table.find('div',class_="rankings-block__banner--name-large")
    Name.append(i.text)
    for i in table.find_all('td',class_="table-body__cell rankings-table__name name"):
        Name.append(i.text.replace('\n',''))

    #scrape position of player
    POS=[]
    i=table.find('span',class_="rankings-block__pos-number")
    POS.append(i.text.replace('                            ','').split('\n')[1])
    for i in table.find_all('span',class_="rankings-table__pos-number"):
        POS.append(i.text.replace('                                    ','').split('\n')[1])
           
    #scarpe the Team of individual player
    Team=[]
    i =soup.find('div',class_="rankings-block__banner--nationality")
    Team.append(i.text.replace('                    ','').split('\n')[2])
    for i in soup.find_all('span',class_="table-body__logo-text"):
        Team.append(i.text)

    #scarpe the Rating of individual player
    Rating=[]
    i =soup.find('div',class_="rankings-block__banner--rating")
    Rating.append(i.text)
    for i in soup.find_all('td',class_="table-body__cell rating"):
        Rating.append(i.text)
    
    #converting all data into dataframe 
    df = pd.DataFrame({'POS':POS,'Name':Name,'Team':Team,'Rating':Rating},index=POS)
    return df.head(10)
```

### 6) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
#importing necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def Coreyms(url):
    
    #query request to website in order to fetch information
    page = requests.get(url)
    
    #download the content
    soup = BeautifulSoup(page.content)

    #get the title
    soup.title.text
    
    #scrape the details 
    #heading
    Heading=[]
    for i in soup.find_all('h2',class_="entry-title"):
        Heading.append(i.text)

    #Date 
    Date=[]
    for i in soup.find_all('time',class_="entry-time"):
        Date.append(i.text)
    
    #content
    Content=[]
    for i in soup.find_all('div',class_="entry-content"):
        Content.append(i.text.replace('\n',''))

    VidCode=[]  
    for i in soup.find_all('iframe',class_="youtube-player"):        
        VidCode.append(i['src'].replace('https://www.youtube.com/embed/','').split('?')[0]) 
      
    
    df = pd.DataFrame({'Heading':Heading,'Date':Date,'Content':Content},index=range(1,11))
    df1=pd.DataFrame({'VidCode':VidCode},index=range(1,10))
    print(df1)
        
    return df
```

### 7) Write a python program to scrape house details from mentioned URL. It should include house title, location,
### area, EMI and price from nobroker.in.


```python
#importing necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def HouseDetails(url):
    #query request to website in order to fetch information
    page = requests.get(url)
    
    #download the content
    soup = BeautifulSoup(page.content)

    #get the title
    soup.title.text
    
    #scrape the details
    #House Title
    Title=[]
    for i in soup.find_all('h2',class_="heading-6 font-semi-bold nb__25Cl7"):
        Title.append(i.text)
        
    Location=[]
    for i in soup.find_all('div',class_="nb__1EwQz"):
        Location.append(i.text.split(',')[1:]) #

    Area=[]
    for i in soup.find_all('div',class_="nb__FfHqA"):
        Area.append(i.text)

    emi=[]
    price=[]
    for i in soup.find_all('div',class_="font-semi-bold heading-6"):
        emi.append(i.text)
        EMI = emi[1::3]
     
        price.append(i.text)
        Price = price[2::3]

    #creating a Dataframe
    df= pd.DataFrame({'Title':Title,'Location':Location,'Area':Area,'EMI':EMI,'Price':Price},index=range(1,17))
    
    return df
```

### 8) Write a python program to scrape mentioned details from dineout.co.in :
#### i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
#import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def Dineout(url):
    #get the request
    page=requests.get(url)
    
    #download the content
    soup = BeautifulSoup(page.content)
    print(soup.title.text)

    #scrape the details from website: Restaurant Name 
    RestaurantName=[]
    for i in soup.find_all('a',class_="restnt-name ellipsis"):
        RestaurantName.append(i.text)
    
    #Cuisine 
    Cuisine=[]
    for i in soup.find_all('span',class_="double-line-ellipsis"):
        Cuisine.append(i.text.split('|')[1])
    
    #Location of Hotel
    Location=[]
    for i in soup.find_all('div',class_="restnt-loc ellipsis"):
        Location.append(i.text)

    #Ratings of hotel
    Rating=[]
    for i in soup.find_all('div',class_="restnt-rating rating-4"):
        Rating.append(i.text)
    
    #image link
    ImageURL=[]
    for i in soup.find_all('img',class_="no-img"):
        ImageURL.append(i['data-src'])

    #converting all the details into Dataframe using Pandas
    df= pd.DataFrame({'Restaurant Name':RestaurantName,'Cuisine':Cuisine,'Location':Location,'Rating':Rating,'ImageURL':ImageURL},index=(range(1,14)))
                     
    return df
```

```python
#link of website
link='https://www.dineout.co.in/delhi-restaurants/buffet-special'
#call the function
Dineout(link)
```

###  9) Write a python program to scrape weather details for last 24 hours from Tutiempo.net :
#### i) Hour ii) Temperature iii) Wind iv) Weather condition v) Humidity vi) Pressure

```python
#import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def Tutiempo(url):
    #get the request
    page=requests.get(url)
    
    #download the content
    soup = BeautifulSoup(page.content)
    print(soup.title.text)

    #scrape the details from website:
    Hour=[]
    for i in soup.find_all('td')[18::6]:
        Hour.append(i.text)
        
    Weather=[]
    for i in soup.find_all('td')[19::6]:
        Weather.append(i.text)
    
    Temperature=[]
    for i in soup.find_all('td')[20::6]:
        Temperature.append(i.text)
        
    Wind=[]
    for i in soup.find_all('td')[21::6]:
        Wind.append(i.text)
        
    Humidity=[]
    for i in soup.find_all('td')[22::6]:
        Humidity.append(i.text)
    
    Pressure=[]
    for i in soup.find_all('td')[23::6]:
        Pressure.append(i.text)
        
    #convert all the details scaraped into dataframe
    df = pd.DataFrame({'Hour':Hour,'Temperature':Temperature,'Wind':Wind,'Weather':Weather,'Humidity':Humidity,'Pressure':Pressure}, index=range(1,50))
    
    return df
```

```python
link='https://en.tutiempo.net/delhi.html?data=last-24-hours'
Tutiempo(link)
```

### 10) Write a python program to scrape monument name, monument description, image URL about top 10 monuments from puredestinations.co.uk

```python
#import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

def Puredestinations(url):
    #get the request
    page=requests.get(url)
    
    #download the content
    soup = BeautifulSoup(page.content)
    print(soup.title.text)

    #scrape the details from website:
    MonumentName=[]
    for i in soup.find_all('p')[4:32:3]:
        MonumentName.append(i.text)
    
    Description=[]
    for i in soup.find_all('p')[5:33:3]:  
        Description.append(i.text)
        
    imageURL=[]
    for i in soup.find_all('img')[6:25:2]:
        imageURL.append(i['src'])
        
    #convert details into dataframe
    df = pd.DataFrame({'MonumentName':MonumentName,'Description':Description,'imageURL':imageURL},index=(range(1,11)))
    
    return df
    
```
