## 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
 Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 
 You need to find following details:
 
 A) Rank B) Name C) Artist D) Upload date E) View

```python
#getting the webdriver
driver = webdriver.Chrome(r"C:/Users/atoranga/Documents/chromedriver.exe")
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)

#creating empty lists to scrape data
name=[]
Rank=[]
Artist=[]
Upload_date=[]
Views=[]

#scraping rank details 
try: 
    tag= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in tag:
        Rank.append(i.text)
        if len(Rank)>29:
            break
except NoSuchElementException as e:
    Rank.append('-')
    
#getting name of videos    
try: 
    tag= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in tag:
        name.append(i.text)
        if len(name)>29:
            break
except NoSuchElementException as e:
    name.append('-')
    
#getting artist name of videos
try: 
    tag= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in tag:
        Artist.append(i.text)
        if len(Artist)>29:
            break
except NoSuchElementException as e:
    Artist.append('-')
    
#scraping views of videos
try: 
    tag= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in tag:
        Views.append(i.text)
        if len(Views)>29:
            break
except NoSuchElementException as e:
    Views.append('-')
    
#scraping the uploaded date of video
try: 
    tag= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in tag:
        Upload_date.append(i.text)
        if len(Upload_date)>29:
            break
except NoSuchElementException as e:
    Upload_date.append('-')
    
#converting all data into dataframe
df = pd.DataFrame({'Rank':Rank,'Name':name,'Artist':Artist,'Upload Date':Upload_date,'Views':Views}, index=Rank)
df
```

## 2. Scrape the details team India’s international fixtures from bcci.tv.

A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#opening the chrome webdriver and loading the url
driver = webdriver.Chrome(r"C:/Users/atoranga/Documents/chromedriver.exe")
url='https://www.bcci.tv/'
driver.get(url)
time.sleep(5)
search=driver.find_element_by_xpath('/html/body/nav/div/div[2]/ul[1]/li[2]/a').get_attribute('href')

driver.get(search)
time.sleep(5)

#creating the empty lists
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]

#scraping titles of match
tag= driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')
for i in tag:
    Match_title.append(i.text.split("-")[0])
       
#scraping Series of match
tag= driver.find_elements_by_xpath('//span[@class="ng-binding"]')
for i in tag:
    Series.append(i.text)
    
#getting place of match held
tag= driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')
for i in tag:
    Place.append(i.text.split("-")[1])

#scraping date of match
tag= driver.find_elements_by_xpath('//div[@class="match-card-left match-schedule"]')
for i in tag:
    Date.append(i.text)
    
#time of match
tag= driver.find_elements_by_xpath('//div[@class="match-card-right match-schedule "]')
for i in tag:
    Time.append(i.text)
    
#converting the match details into dataframe
df = pd.DataFrame({'Match title':Match_title, 'Series':Series, 'Place':Place, 'Date':Date,'Time':Time})
df
```

```python
#opening the chrome driver
driver = webdriver.Chrome(r'C:/Users/atoranga/Documents/chromedriver.exe')
url = 'https://github.com/'
driver.get(url)
time.sleep(2)
```

```python
#opening the link and sliding on explore tab
tag=driver.find_elements_by_xpath('//details[@class="HeaderMenu-details details-overlay details-reset width-full"]')
for i in tag:
    if i.text == 'Explore':
        driver.find_element_by_xpath('//details[@class="HeaderMenu-details details-overlay details-reset width-full"]').click
     
```

```python
 #clicking on the trending repositories
tag=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').get_attribute('href')
driver.get(tag)
time.sleep(2)
```

```python
#creating the empty lists for data to be scraped
title=[]
desc=[]
Contributors=[]
Language=[]
links=[]
lang=[]

#scraping title details
try:
    tag = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a')
    for i in tag:
        title.append(i.text.split('/')[1])   
except:
    title.append('-')

#scraping contributor details
try:
    tag = driver.find_elements_by_xpath('//span[@class="d-inline-block mr-3"]')
    for i in tag:
        Contributors.append(len(i.find_elements_by_tag_name('a')))
         
except:
    Contributors.append('-')

#getting into each links and storing into link list.
tags = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a')
for i in tags:
    links.append(i.get_attribute('href'))

for j in links:
    driver.get(j)
    #scraping languages used in each repository 
    try:
        tag = driver.find_elements_by_xpath('//span[@class="color-fg-default text-bold mr-1"]')
        for i in tag:
            lang.append(i.text)
        Language.append(lang)   
    except:
        Language.append('-')
    #scraping description of each repository
    try:
        tag = driver.find_element_by_xpath('//p[@class="f4 mb-3"]')
        desc.append(tag.text)        
    except NoSuchElementException:
        desc.append('-')        
    
    
        
#conveting all details into dataframe
df = pd.DataFrame({'Repository title':title, 'Repository description':desc ,'Contributors count':Contributors, 'Language used':Language})
df
```
